```python
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from konlpy.tag import Okt
import requests
import time
import pandas as pd
import numpy as np
import re
```

```python
df_raw = pd.read_csv('Card_Data.csv')
df_raw.head()
```

```python
# 정규식을 이용한 문자열 검색

re_option = []

for row in df_raw['카드명']:
    
    row = re.sub("\n", "", row)
  
    re_option.append(row)
```

```python
re_option
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

option_txt = []

for opt in re_option:

    driver = webdriver.Chrome('./chromedriver')
    driver.get('https://www.shinhancard.com/pconts/html/card/main/main.html')
    time.sleep(2)

    driver.find_element(By.XPATH, '//*[@id="header"]/div[1]/div[3]/div/div[3]/button[1]').click()
    time.sleep(1)

    element = driver.find_element(By.XPATH, '//*[@id="totalSrchQuck"]')
    element.send_keys(opt)
    element.send_keys(Keys.RETURN)

    time.sleep(5)

    
    check = len(driver.find_elements(By.XPATH,'//*[@id="content0"]/div/div[2]/div[1]/ul/li[1]/div[2]/a/img' ))

    if check != 0:
        yn = driver.find_element(By.XPATH, '//*[@id="content0"]/div/div[2]/div[1]/ul/li[1]/div[2]/a/img')
        yn.click()

        time.sleep(3)

        check1 = len(driver.find_elements(By.XPATH, '//*[@id="cardCompareAfter"]/div[3]/div/dl/dt[3]/button'))
        if check1 != 0:

            driver.find_element(By.XPATH, '//*[@id="cardCompareAfter"]/div[3]/div/dl/dt[3]/button').send_keys(Keys.ENTER)
            time.sleep(1)

            tt = driver.find_element(By.XPATH, '//*[@id="acc2"]').text

            option_txt.append(tt)
        else:
            option_txt.append(' ')
            continue

    
    else:
        option_txt.append(' ')
        continue
```

```python
df_option = pd.DataFrame({'option_crawl' : []})
df_option['option_crawl'] = option_txt
```

```python
df_option.to_csv('df_option.csv')
```
