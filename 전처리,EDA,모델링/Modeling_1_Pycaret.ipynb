```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# from pycaret.classification import *
```

```python
import pycaret
from pycaret.classification import *
```

```python
data = pd.read_csv('TM_meta.csv')
del data['연령']
del data['직업군']
del data ['학력']
del data['contact']
del data['연락일']
del data['요일']
del data['주말여부']
del data['년']
del data['일']
del data['Card_ID']
del data['소개']
del data['옵션']
del data['카드명']
data.head()
```

```python
data['계약여부'] = np.where(data['계약여부'] == 'no', 0, 1)
```

```python

df_data = pd.get_dummies(data)
df_data.head()
```

```python

numeric_features = ['과거통화횟수', '연령대', '월', '후불교통가능', '가족카드가능', '현금카드가능', 'IC기능']
```

Setup Function

```python
# 성능 제일 좋았던 것

b_setup = setup(data= df_data, 
            target='계약여부', 
            numeric_features=numeric_features,         
            normalize=True,                            # 정규화 적용
            normalize_method='zscore',              
            train_size = 0.7,
            fix_imbalance = True, 
            fold = 100,
            pca = True,
            pca_method = 'kernel',
            session_id= 1234
            ) 
```

Compare_Models

```python
# fold 100

top3_model_fold100 = compare_models(
               round=4,
               sort="F1",
               n_select = 3 )
```

Tuning

```python
tuned_top3 = [tune_model(i,choose_better = True, optimize = 'F1', n_iter = 10) for i in top3_model_fold100 ]
```

Blending

```python
blender_specific = blend_models(estimator_list = top3_model_fold100, method = 'auto',  choose_better = True)
```

Stacking

```python
stacker = stack_models(estimator_list = [xgboost, lgbm], meta_model = catboost)
```
